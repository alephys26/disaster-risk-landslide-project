```python
import pandas as pd
data = pd.read_csv('landslide_data.csv')
data['co'] = list(zip(data['latitude'].astype(int),data['longitude'].astype(int)))
data
```

```python
records = {}
for i, val in data.iterrows():
    if val['co'] not in records:
        records[val['co']] = [i]
    else:
        records[val['co']] += [i]
print(records)
```

```python
import os
file_temp = "./dem_data/row_"
for key, item in records.items():
    for c in range(2, len(item)):
        file = file_temp + f"{item[c]}.csv"
        os.remove(file)
```

```python
file_temp = "./dem_data/row_"
dem_data = {'Index':[],
        'Slope':[], 'Aspect':[], 'Plan Curvature':[], 'Profile Curvature':[], 
        'TPI':[], 'Elevation Relative':[], 'Elevation Percentile':[]
    }

for key, item in records.items():
    dem = pd.read_csv(file_temp+f"{item[0]}.csv")
    for value in item:
        row = round((data.loc[value, "latitude"] - key[0])*3600)
        col = round((data.loc[value, "longitude"] - key[1])*3600)
        dem_value = dem[(dem['Row'] > row - 2) & (dem['Row'] < row + 2) & 
                (dem['Col'] > col - 2) & (dem['Col'] < col + 2)]
        dem_value = dem_value.mean()
        dem_value.drop(['Row', 'Col'], inplace=True)
        for i, j in dem_value.iteritems():
            dem_data[i].append(j)
        dem_data['Index'].append(value)
    del dem
```

```python
dem_df = pd.DataFrame(dem_data)
dem_df.set_index('Index', inplace=True)
dem_df.sort_index(inplace=True)
dem_df.reset_index(inplace=True)
dem_df.drop(columns=['Index'], inplace=True)
dem_df
```

```python
dem_df.to_csv('dem_data.csv')
```
