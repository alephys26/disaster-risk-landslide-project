```python
def process_rainfall(i, data):
    df = pd.read_csv(f'imerg_data/row_{i}.csv')
    df = df[df['nv']!=1]
    df.drop(columns=['precipitation_cnt',
    'MWprecipitation', 'MWprecipitation_cnt', 'MWprecipitation_cnt_cond',
    'randomError', 'randomError_cnt', 'lon', 'lat', 'time_bnds', 'probabilityLiquidPrecipitation'], inplace=True)
    df.reset_index(inplace=True)
    df = df[df['nv']!=1]
    df.drop(columns=['nv'], inplace=True)
    df = df.transpose()
    df.drop(['index', 'time'], axis=0, inplace=True)
    df.columns = [f"crh{i}" for i in range(1,11)]
    new_cols = [f"rf{i}" for i in range(1,11)]
    row = df.loc['precipitation']
    row2 = df.loc['precipitation_cnt_cond']
    row.index = new_cols
    merged_row = pd.concat([row2, row])
    return data.append(merged_row, ignore_index=True)

columns = [f"crh{i}" for i in range(1,11)] + [f"rf{i}" for i in range(1,11)]
rdict = {i:[] for i in columns}
rain_data = pd.DataFrame(rdict)
for i in range(263):
    rain_data = process_rainfall(i, rain_data)
rain_data
```

```python
rain_data.to_csv('rainfall_data.csv')
```
